```python
import pandas as pd
import numpy as np
import cirpy
from cirpy import Molecule
```

```python
inventory = pd.read_csv('/Users/Emily/Desktop/NP_casids.csv', index_col=0)
inventory.head()
```

```python
def add_smiles(casid):
    casid = str(casid)
    smiles = cirpy.resolve(casid, 'smiles')
    return smiles
```

```python
inventory['SMILES code'] = inventory.apply(lambda row: add_smiles(row.CASID), axis=1)
inventory.head()
```
